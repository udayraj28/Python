A **loop in Python** is a control structure that allows a block of code to be executed repeatedly as long as a condition is true or for each item in a sequence.
Two Types of loops are there in python
1. for loop
2. while loop

# for loop

```python
name = "Uday"
for i in name:
    print(i)
```

```python
dct = {"name":"uday", "class":"3rd year"}
for items in dct:
    print(items)
```

```python
for items in dct.values():
    print(items)
```

```python
for items in dct.items():
    print(items)
```

```python
for i in range(2, 22, 2):
    print(i)
```

```python
lst = [1,2,3,4,5,6,7,8,9]
even = []
odd = []
for i in lst:
    if i % 2 == 0:
        print(f"{i} is even")
        even.append(i)

    elif i % 2 != 0:
        print(f"{i} is odd")
        odd.append(i)
    
    else:
        pass
print("even numbers in list",even)
print("odd numbers in list",odd)
```

# While Loop

```python
i = 10

while i <= 15:
    print(i, end= " ")
    i += 1
```

```python
i = 1

while i <= 5:
    print(f"Cube of {i} is:", i**3)
    i += 1
```

```python
i = 1

while i <= 10:
    if i % 2 != 0:
        print(i, end=" ")
    i += 1
```

```python
i = 1
while i <= 5:
    print("*" * i)
    i += 1
```

```python
i = 1
mul = 1
while i <= 5:
    mul = mul * i
    i += 1
print(mul)
```
