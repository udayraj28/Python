# Lambda function
A lambda function in Python is a small anonymous (nameless) function that can have any number of arguments but only one expression.

Syntax
lambda arguments: expression

```python
square = lambda x: x * x

print(square(5))
```

```python
def max_of_three(a, b, c):
    if a > b and a > c:
        print(f"{a} is Greater")
    elif b > a and b > c:
        print(f"{b} is Greater")
    else:
        print(f"{c} is greater")

max_of_three(5, 6, 9)
```

```python
def count_vowels(s):
    lst = ["a", "e", "i", "o", "u", "A", "E", "I", "O", "U"]
    count = 0

    for i in s:
        if i in lst:
            print(f"{i} is vowel")
            count += 1
    print("Count of Vowels are:", count)

count_vowels("Hello, World!")
```

```python
def calculator(a, b, op):
    if op == "+":
        print(f"sum is: ", a + b)
    elif op == "-":
        print(f"Subtraction is: ", a - b)
    elif op == "*":
        print(f"Multiplication is:", a * b)
    elif op == "/":
        print(f"Division is: ", a / b)
    else:
        pass

calculator(5, 2, "/")
```

```python
def sum(*args):
    print(*args)
    total = 0
    for i in args:
        total = total + i
    
    print("sum of all number is:", total)

sum(10,20,30)
    
```

```python
def factorial(n):
    if n == 0 or n == 1: 
        return 1
    return n * factorial(n - 1)

factorial(5)
```

```python
def apply_twice(func, value):
    return  func(func(value))

def double(x):
    return x * 2

apply_twice(double, 3)
```
