A conditional statement in Python is a statement that lets a program make decisions by executing different code depending on whether a condition is True or False.

The most common conditional statements are if, elif, and else.

```python
age = int(input("Enter your Age"))
if age > 18:
    print("you can give vote 😊")

else:
    print("you can not give vote 😭")
```

```python
age = int(input("Enter your Age"))
if age > 18:
    print("you can give vote 😊")
elif age == 18:
    print(f"Aap vote de sakte ho✌️ aapki age hai {age}")
else:
    print("you can not give vote 😭")
```

```python
a = int(input("Enter a number:"))

if a % 2 == 0:
    print(f"Number {a} is even")

else:
    print(f"Number {a} is odd")
```

```python
# program for grading system
num = int(input("Enter a number score out of 100"))

if 91 <= num >= 100:
    print("Your Grade is A++")
elif 81 <= num >= 90:
    print("Your Grade is A+")
elif 71 <= num >= 80:
    print("Your Grade is B++")
elif 61 <= num >= 70:
    print("Your Grade is B")

else:
    print("Your grad is C")
```

```python
# program to find discount on purchase

purchase = int(input("Enter a total amount of purchase in Rupess:"))
print(f"Total Purchase is {purchase}")
if purchase > 5000:
    discount = purchase*(20/100)
    print(f"total discount is {discount}")
elif purchase > 2000:
    discount = purchase*(10/100)
    print(f"total discount is {discount}")
else:
    print(f"Not eligible for discount")
```
