File handling in Python is used to create, read, write, append, and manage files stored on a computer. It allows data to be stored permanently and accessed whenever needed.

Python uses the open() function to open a file in different modes such as:

r – Read
w – Write
a – Append
x – Create a new file

After opening a file, operations like reading (read()), writing (write()), and appending data can be performed. Files should be closed using close() after use, or managed automatically using the with statement.

File handling helps in storing program data, maintaining records, generating reports, and working with both text and binary files efficiently.

```python
file = open("new.txt", 'x') # create file new.txt
file.write("hello python")
file.close()
```

```python
file = open("new.txt", 'r')
print(file.read())
file.close()
```

```python
file = open("new.txt", 'w')
file.write("Hii Sunil !!")
file.close()

file = open("new.txt", 'r')
print(file.read())
file.close()
```
